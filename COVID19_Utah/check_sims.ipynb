```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import glob
from pathlib import Path
data_N= []
data_K = []




comp = 'cumI'

for filename in Path('model_output/low-utah_NoNPI/').rglob('*.csv'):

    sim = pd.read_csv(filename)
    sim = sim[sim['comp']==comp]
    sim.drop('comp', axis =1, inplace=True)
    sim.set_index('time', inplace = True)
    sim = sim.sum(axis = 1)
    data_N.append(sim)


for filename in Path('model_output/mid-utah_NoNPI/').rglob('*.csv'):
    sim = pd.read_csv(filename)
    sim = sim[sim['comp']==comp]
    sim.drop('comp', axis =1, inplace=True)
    sim.set_index('time', inplace = True)
    sim = sim.sum(axis = 1)
    data_W.append(sim)
    


data_N = pd.concat(data_N,axis=1)
data_W = pd.concat(data_W,axis=1)
```

```python
fig, axes = plt.subplots(1,1, figsize =(10,6))


axes.plot(data_N.quantile(.5, axis = 1), c = 'r')
axes.fill_between(data_N.index, data_N.quantile(.05, axis = 1), 
                                data_N.quantile(.95, axis = 1), alpha =.1, 
                                facecolor ='r', label = 'Nothing')

axes.plot(data_W.quantile(.5, axis = 1), c = 'r')
axes.fill_between(data_N.index, data_W.quantile(.05, axis = 1), 
                                data_W.quantile(.95, axis = 1), alpha =.1, 
                                facecolor ='r', label = 'Nothing')





#axes.set_xlim(50,200)
axes.legend()
#axes.set_ylim(0,500000)
fig.autofmt_xdate()
```
