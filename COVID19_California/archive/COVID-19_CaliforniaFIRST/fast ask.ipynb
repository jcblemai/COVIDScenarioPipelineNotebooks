```python
import pandas as pd
import numpy as np
import datetime

a = []
for i in range(1000):
    sim = pd.read_csv(f'model_output/SanFrancisco/mid/sim_{i}_scn_midSF.csv')
    sim = sim[sim['comp']=='diffI']
    a.append(int(sim[sim['time']=='2020-03-09'].sum(axis=1)))
```

```python
print('mean', np.mean(a), 'median', np.median(a), '20th',np.quantile(a, .2), '80th', np.quantile(a, .8))
```
