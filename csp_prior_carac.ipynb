```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ndraw = int(1e6)
nnpi=6
true_npis_val = .25
# say true NPI as a value of .25 and there 6 of them
effective_reduction = np.ones((ndraw))
#scipy.stats.truncnorm((a - mean) / sd, (b - mean) / sd, loc=mean, scale=sd)

for i in range(nnpi):
    effective_reduction *= (1-np.random.uniform(true_npis_val-.2,true_npis_val+.2, (ndraw)))
    #effective_reduction *= (1-np.random.normal(loc =true_npis_val, scale=0.2, size=ndraw))

effective_reduction = 1-effective_reduction
mean_reduction = 1-(1-true_npis_val)**nnpi

fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.hist(effective_reduction, bins=100)
ax.vlines(x=mean_reduction, ymin=0,ymax = ndraw/40, lw =4, color ='r')
ax.vlines(x=np.mean(effective_reduction), ymin=0,ymax = ndraw/60, lw =3, color ='lightgreen')
ax.vlines(x=np.median(effective_reduction), ymin=0,ymax = ndraw/60, lw =3, color ='orange')
ax.set_xlim(mean_reduction - .3,mean_reduction + .3)
ax.grid()
```
