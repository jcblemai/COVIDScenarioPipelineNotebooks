### Arithmetic rate

```python
geoids = ['1001', '2020', '3030'] 
affected = ['1001', '3030'] 
start_date = datetime.date(2020, 3,1)
end_date = datetime.date(2020, 6,1)
period_start_date = datetime.date(2020, 3, 15)
period_end_date = datetime.date(2020, 5, 15)

npi = pd.DataFrame(0.0, index=geoids,
                                    columns=pd.date_range(start_date, end_date))
dist = [.5,.3]#npi_config["value"].as_random_distribution()
period_range = pd.date_range(period_start_date, period_end_date)

rate = np.array([.1, .9])
freq_days = 28
min_value = .1
npi.loc[affected, period_range] = (np.tile(dist*np.ones(len(affected)), (len(period_range), 1)).T  - \
                                   np.tile(dist*np.ones(len(affected)), (len(period_range), 1)).T * \
                                  np.tile(rate, (len(period_range), 1)).T*(np.arange(0,len(period_range))/freq_days)).clip(min_value)
                                          
                          
```
